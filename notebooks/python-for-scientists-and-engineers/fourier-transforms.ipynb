```python
from scipy import *
from scipy import fftpack
import matplotlib.pyplot as plt
import numpy as np
import numpy.fft as fft


signal_time = 0.5    # signal length [s]
sample_rate = 500    # sampling rate [Hz]

dt = 1 / sample_rate    # time between samples [s]
df = 1 / signal_time    #  [Hz]

# array for time
t = np.arange(0, signal_time, dt)

# positive frequencies
freq = np.arange(0, sample_rate/2, df)
n = len(freq)   # length of frequency array

# generate the signal
y = 0.5 * np.sin(2 * np.pi * 40 * t) + np.sin(2 * np.pi * 60 * t + np.pi/2)

f = fft.fft(y)      # compute the fourier transform

# plotting signal
fig, ax = plt.subplots(2, 1)
ax[0].plot(t, y)
ax[0].set_title('time domain')
ax[0].set_xlabel('time [s]')

# plotting the fourier transform
ax[1].plot(freq, abs(f[0:n]))       # 
ax[1].set_title('FFT')
ax[1].set_xlabel('frequency [Hz]')
ax[1].set_ylabel('abs(DFT)')

fig.tight_layout()
```
