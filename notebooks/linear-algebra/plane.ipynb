Plot a plane in R3 from two basis vectors

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Import needed for 3D plotting

# Define a point on the plane and the normal vector to the plane
point = np.array([1, 0, 1/2])
normal = np.array([0, 1, 1/2])

# Calculate d using the plane equation: a*x + b*y + c*z + d = 0
d = -point.dot(normal)

# Create a grid of x, y values
xx, yy = np.meshgrid(range(10), range(10))

# Calculate the corresponding z values for the plane
z = (-normal[0] * xx - normal[1] * yy - d) / normal[2]

# Set up the figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the plane surface
ax.plot_surface(xx, yy, z, alpha=0.5, rstride=1, cstride=1, cmap='viridis')
ax.view_init(elev=30, azim=60)  # Adjust these values as needed
# Set labels for axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot
plt.show()
```
