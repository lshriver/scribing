# ⚠️ Gradient Vector Field is not Properly Rendered

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from numpy import *
import numpy as np

# Increasing the resolution of the plot
plt.figure(dpi=100)

# Activating text rendering by LaTex
# plt.rcParams.update({
#    "text.usetex": True,
#    "font.family": "monospace",
#    "font.monospace": 'Computer Modern Typewriter'
# })
```

```python
def f(x:float, y: float) -> float:
    return (x**2) + (y**2)

x = np.linspace(-2, 2, 30)
y = np.linspace(-2, 2, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

figure = plt.figure(figsize=(5,5))
ax = figure.add_subplot(111, projection='3d')
surface = ax.plot_surface(X, Y, Z,  cmap='Blues', edgecolor='none', antialiased=False, alpha=.7)

u, v= np.meshgrid(np.linspace(-2,2,15), np.linspace(-2,2,15))
z = f(u, v)
U,V = np.gradient(z)
Z0 = np.zeros_like(u)-1

ax.quiver(u, v, Z0, u+U, v+V, Z0, linewidth=1, length=0.1, color='darkorange')
plt.title(r'$f(x,y)=x^2+y^2$ and $\vec\nabla f(x,y)$')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')

plt.xticks(np.arange(min(x), max(x)+1, 1.0))
plt.yticks(np.arange(min(y), max(y)+1, 1.0))
ax.set_zlim(-2, 8)
plt.show()
```
