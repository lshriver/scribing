# Plotting a function

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x: float) -> float:
    return x**4                 # define the function

inputs = np.linspace(-11, 10, 45)
outputs = list(map(f, inputs))

fig = plt.figure()
ax = fig.gca()
plt.plot(inputs, outputs, '.')
plt.title(r'$f(x)=x^2$')
plt.xlabel(r'$x$')
plt.grid()
plt.show()
```
