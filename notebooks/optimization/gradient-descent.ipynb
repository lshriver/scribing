# ⚠️ Gradient Vector Field is not Properly Rendered

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from numpy import *
import numpy as np
```

```python
def f(x:float, y: float) -> float:
    return (x**2) + (y**2)

x = np.linspace(-2, 2, 30)
y = np.linspace(-2, 2, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

figure = plt.figure(figsize=(10,10))
ax = figure.add_subplot(111, projection='3d')
surface = ax.plot_surface(X, Y, Z,  cmap='Blues', edgecolor='none', antialiased=False, alpha=.7)



def grad_f(x: float, y: float) -> tuple[float, float]:
    '''
    Input a point in the plane and outputs the gradient at that point
    '''
    return np.array([2*x + y**2 , x**2 + 2*y])

U,V = grad_f(X,Y)
Z0 = np.zeros_like(X)
ax.quiver(X, Y, Z0, X+U, Y+V, Z0, normalize=True, linewidth=0.6, length=0.15, color='darkorange')
plt.show()
```
