# Qubit Representation using `qutip`

## Background
Suppose you have a vector $\vec{r}=\begin{bmatrix}x \\ y \\ z\end{bmatrix}$ on the Bloch sphere. 
- Note: Since $\vec{r}$ is on the Bloch sphere, we can assume it has a magnitude of 1.
- Cartesian $\rightarrow$ polar coordinates:
  
  $$x=\sin{\theta}\cos{\phi}, \\ y=\sin{\theta}\sin{\phi} \\ z=\sin{\phi}$$

The corresponding state will be
$$\ket{\psi}=\cos{\bigg( \frac{\theta}{2} \bigg)}\ket{0} + e^{i\phi}\sin{\bigg( \frac{\theta}{2} \bigg)}\ket{1}$$

We can write this state vector 'in computational basis' as follows
$$\ket{\psi}=\begin{bmatrix} \cos{\big( \frac{\theta}{2} \big)} \\ \\e^{i\phi}\sin{\bigg( \frac{\theta}{2} \bigg)}\end{bmatrix}$$

```python
import numpy as np
import matplotlib.pyplot as plt 
from qutip import *
```

```python
num_qubits = 1
state = rand_ket(2**num_qubits)

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
sphere = Bloch(axes=ax)
sphere.add_states(state)
sphere.show()
```

```python
vec = [0,1,0]

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
sphere = Bloch(axes=ax)
sphere.add_vectors(vec)
sphere.show()
```

```python
vec = [0,1,0]
up = basis(2,0)

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
sphere = Bloch(axes=ax)
sphere.add_vectors(vec)
sphere.add_states(up)
sphere.show()
```

```python
x = (basis(2,0)+(1+0j)*basis(2,1)).unit()
y = (basis(2,0)+(0+1j)*basis(2,1)).unit()
z = (basis(2,0)+(0+0j)*basis(2,1)).unit()


fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
sphere = Bloch(axes=ax)
sphere.add_states([x,y,z])
sphere.show()
```

```python
theta = np.pi/3
phi = (5*np.pi)/6
vec = [np.sin(theta)*np.cos(phi),np.sin(theta)*np.sin(phi),np.cos(theta)]
# state = (basis(2,0)+(np.sqrt(3)/2 + 0j)*basis(2,1)+(-np.sqrt(3)/4 + (1/4)*1j)).unit()
# z = (basis(2,0)+(0+0j)*basis(2,1)).unit()


fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
sphere = Bloch(axes=ax)
custom_colors = ['#1572ea', 'purple', 'yellow', 'green']
sphere.vector_color = custom_colors
sphere.add_vectors(vec)
# sphere.add_states(state)
sphere.show()
```

Add a set of 27 points around the equator ($xy$-plane)

```python
theta = np.linspace(0, 2*np.pi, 27)
x_xy = np.cos(theta)
y_xy = np.sin(theta)
z_xy = np.zeros(27)

pnts = [x_xy, y_xy, z_xy]

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
b = Bloch(axes=ax)
b.add_points(pnts, 'm') # <-- add a 'm' string to signify 'multi' colored points
b.show()
```

Add a set of 27 points around the $yz$-plane

```python
x_yz = np.zeros(27)
y_yz = np.sin(theta)
z_yz = np.cos(theta)

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
b = Bloch(axes=ax)
b.add_points([x_yz, y_yz, z_yz])
b.render()
b.show()
```
