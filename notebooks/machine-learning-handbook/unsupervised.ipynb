# Unsupervised Learning

## $k$-Means Clustering

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

X = np.array([[1, 2], [1.5, 9], [2, 2], [5, 6], [8, 8], [3, 8], [9, 10], [3, 3], [4,2]])
clustering = KMeans(n_clusters=3).fit(X)
labels = clustering.labels_
colors = ("red","green","blue")
plt.figure(figsize=(8, 6))
for i in range(len(X)):
  plt.scatter(X[i][0],X[i][1], c = colors[labels[i]], marker = 'x')
```

## PCA

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Load the iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Apply PCA to reduce the data to 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot the results
plt.figure(figsize=(8, 6))

for i in range(len(set(y))):
  plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], label=f'Class {i}')

plt.title('PCA of Iris Dataset')
plt.xlabel('Principle Component 1')
plt.ylabel('Principle Component 2')
plt.legend()
```
